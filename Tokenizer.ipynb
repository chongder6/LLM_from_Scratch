```python
#this code is for opening a text file and reading it first the address
#then the r which means reading ||| then the encoding which is important else
#you will be getting error ||| then the raw_text which holds the book

with open("/content/the-verdict.txt","r", encoding="utf-8") as f:
  raw_text = f.read()
print("Total number of characters: ", len(raw_text)) #counting the number of chars
print(raw_text[:99]) #for illustration purpose
```

```python
#now splitting the text
#re is for regular expressions (like in which basis we will split the text)
#\s is like basis of space i am splitting the total text

import re
text = raw_text
result = re.split(r'(\s)', text)

print(result)
```

```python
#separating on the basis of , and . also with spaces

result = re.split(r'([,.]|\s)', text)
print(result)
```

```python
#removing the whicte spaces using strip
#which is not required (btw this step is optional)
#sometimes like for ex. you have python code for training data then the
#white spaces are important as python is sensetive to indentation
#for now i am removing for memory optimization

result = [item for item in result if item.strip()] # this version only filters out
print(result)
```

```python
#this is the final which everything excluded

prepros = re.split(r'([.,:;?/!"()\']|--|\s)', text)
prepros = [item.strip() for item in prepros if item.strip()] #this version filters out and transform

print(prepros[:20])
```

```python
print(len(prepros))
```

### **Now Tokens are there, we need to create token IDs **

```python
# now we will take the tokens and sort them alphabetically to determine the vocabulary size
#and remember the vocabulary only consist of unique words that is why the size maybe different form the org
all_words = sorted(set(prepros))
vocab_len = len(all_words)

print(vocab_len)
```

```python
#now we need to create the vocab dictionary
# this basically means for every token assign an integer (token:integer)
#and how we are assigning is enumerate all word and alohabetically
#giving all the numbers/ids

vocab = {token:integer for integer, token in enumerate(all_words)}

# print(vocab)
```

```python
#we are using enumerate and all to clean up the code
#else without that we need to write lot of lines ant this also give
#us auto counting and all

for i,item in enumerate(vocab.items()):
  print(item)
  if i>20:
    break
```

```python
# now as we are encoding this when we will be needing the output we want our words back so in that case
#we need a decoder also
#So we will implement a tokenizer class which will have encode method and
#one decode method which will carry out the reverse int-to-str mapping

class SimpleTokenizerV1:
  def __init__(self, vocab):
    self.str_to_int = vocab
    self.int_to_str = {i:s for s,i in vocab.items()}

  def encode(self, text):
    prepros = re.split(r'([.,:;?!()"\']|--|\s)', text)
    prepros = [
        item.strip() for item in prepros if item.strip()
        ]
    ids = [self.str_to_int[s] for s in prepros]
    return ids

  def decode(self, ids):
    text = " ".join([self.int_to_str[i] for i in ids])
    #here replacing spaces before the punctuations
    text = re.sub(r'\s+([.,:;?!()"\'])', r'\1', text)
    return text

#vocab - is the mapping for str_to_int it goes like this
#for the int_to_str we are taking the string and the integer in the vocabulary
#then for every integer we are assigning one string or mention which token it is
#inside that ids we are just giving token their token IDs

#in the decoder we are reversing just i.e we are converting token IDs
#into individual token and also joining
```

```python
#here i am creating an instance of the claass basically testing it out

tokenizer = SimpleTokenizerV1(vocab)
text = """it's the last he painted, you know,"
          Mrs. Gisburn said."""
ids = tokenizer.encode(text)
print(ids)
```

```python
#now here you can see the decode method successfully running

tokenizer.decode(ids)

#but the thing is the input which i gave it's from the training set
#but if something came up which is not in the vocabulary
```

## SO WE WILL BE USING SPECIAL CONTEXT TOKENS

Basically we modify the tokenizer to handle unknown words

```python
#here creating the version 2 where we will add end of text and unknown
#two things to remember here----
#1. <|unk|> when to handle a situation where word is not known

#2. <|endoftext|> where multiple sources of data are there and after the first source
#the 2nd is starting then at the end of the first this token will go

#first we will modify the vocab

all_tokens = sorted(set(prepros))
all_tokens.extend(["<|endoftext|>","<|unk|>"])

vocab = {token:integer for integer, token in enumerate(all_tokens)}
```

```python
class SimpleTokenizerV2:
  def __init__(self, vocab):
    self.str_to_int = vocab
    self.int_to_str = {i:s for s, i in vocab.items()}

  def encode(self, text):
    preprocess = re.split(r'([.,:;?!()"\']|--|\s)', text)
    preprocess = [
        item.strip() for item in preprocess if item.strip()
    ]
    #this is the extra line or change we have do rest is same

    preprocess = [
        item if item in self.str_to_int
          else "<|unk|>" for item in preprocess
    ]
    ids = [self.str_to_int[s] for s in preprocess]
    return ids

  def decode(self, ids):
    text = " ".join([self.int_to_str[i] for i in ids])
    text = re.sub(r'\s+([.,:;?!()"\'])', r'\1', text)
    return text
```

```python
#now lets try this

new_tok = SimpleTokenizerV2(vocab)

new_text1 = """Hello , here is the grace"""
new_text2 = """ this i am testing for the next which is
                  that" end of text token"""
#here we are using it as we will not combine two texts we will
#separate this instead using the endoftext

merge_text = "<|endoftext|>".join((new_text1, new_text2))
print(merge_text)
```

```python
#same encoding doing without error

mt = new_tok.encode(merge_text)
print(mt)
```

```python
#we are handling the unknown words now
new_tok.decode(mt)
```
